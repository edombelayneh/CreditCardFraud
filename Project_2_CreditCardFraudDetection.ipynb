```python
# change the dob column from date of birth to age by subtracting from current year -> 2024
from datetime import datetime
current_year = datetime.now().year
df_train['dob'] = current_year - pd.to_datetime(df_train['dob']).dt.year
df_test['dob'] = current_year - pd.to_datetime(df_test['dob']).dt.year

df_train
```

```python
# change the title 'dob' to age
df_train = df_train.rename(columns={'dob': 'age'})
df_test = df_test.rename(columns={'dob': 'age'})

df_train
```

```python
# age vs fraud
sns.boxplot(x='is_fraud', y='age', data=df_train)
plt.title('Age vs Fraudulent Transactions')
```

```python
# separate features and target, target -> is_fraud
X_train = df_train.drop('is_fraud', axis=1)
X_test = df_test.drop('is_fraud', axis=1)

y_train = df_train['is_fraud']
y_test = df_test['is_fraud']
```

```python
X_train = X_train.drop(['Unnamed: 0', 'first', 'last', 'street', 'state', 'cc_num', 'trans_date_trans_time', 'zip', 'trans_num', 'city', 'gender'], axis=1)
X_test = X_test.drop(['Unnamed: 0', 'first', 'last', 'street', 'state', 'cc_num', 'trans_date_trans_time', 'zip', 'trans_num', 'city', 'gender'], axis=1)
```

```python
X_train
```

```python
# handle or drop missing values
X_train = X_train.dropna()
X_test = X_test.dropna()

y_train = y_train.dropna()
y_test = y_test.dropna()

X_train
```

```python
# label encoding. this helps from the table getting to wide where it assigns integers to categorical data
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X_train['category'] = le.fit_transform(X_train['category'])
X_train['merchant'] = le.fit_transform(X_train['merchant'])
X_train['job'] = le.fit_transform(X_train['job'])
# X_train['state'] = le.fit_transform(X_train['state'])
# X_train['city'] = le.fit_transform(X_train['city'])

X_test['category'] = le.fit_transform(X_test['category'])
X_test['merchant'] = le.fit_transform(X_test['merchant'])
X_test['job'] = le.fit_transform(X_test['job'])
# X_test['state'] = le.fit_transform(X_test['state'])
# X_test['city'] = le.fit_transform(X_test['city'])


X_train
```

```python
X_test
```

```python
# normalizing or scaling independent variables
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Train Machine Learning Model
# create a method that takes in the model and does the rest

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pickle

def evaluate_and_save(model, X_train, X_test, y_train, y_test, filename):
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)
  accuracy = accuracy_score(y_test, y_pred)
  print(f'{model.__class__.__name__} Accuracy: {accuracy:.4f}')
  print(f'\nClassification Report: \n{classification_report(y_test, y_pred)}')
  print('----------------------------------')

  with open(filename, 'wb') as file:
    pickle.dump(model, file)

  print(f'Model saved as {filename}')
```

```python
# Train Machine Learning Model

# Logistic Regression Model - used for binary classification model
# XGBoost - good for tabular data
# Decision Tree - ask a series of yes or no
# Random Forest - a bunch of Decision Trees but asking for opinion
# Naive Bayes
# K Nearest Neighbors - finding similarity
# Support Vecotr Machines
```

```python
from sklearn.linear_model import LogisticRegression
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
lr_model = LogisticRegression()
xgb_model = xgb.XGBClassifier()
dt_model = DecisionTreeClassifier()
rf_model = RandomForestClassifier()
svm_model = SVC()
nb_model = GaussianNB()
knn_model = KNeighborsClassifier()
```

```python
evaluate_and_save(lr_model, X_train, X_test, y_train, y_test, "lr_model.pkl")
evaluate_and_save(xgb_model, X_train, X_test, y_train, y_test, "xgb_model.pkl")
evaluate_and_save(dt_model, X_train, X_test, y_train, y_test, "dt_model.pkl")
```

```python
# 2. SMOTE - improve predictions by oversampling the minority in the dataset
from imblearn.over_sampling import SMOTE
```

```python
X_train_resampled, y_train_resampled = SMOTE().fit_resample(X_train, y_train)
```

```python
evaluate_and_save(lr_model, X_train_resampled, X_test, y_train_resampled, y_test, "lr_model_resampled.pkl")
evaluate_and_save(xgb_model, X_train_resampled, X_test, y_train_resampled, y_test, "xgb_model_resampled.pkl")
evaluate_and_save(dt_model, X_train_resampled, X_test, y_train_resampled, y_test, "dt_model_resampled.pkl")
# evaluate_and_save(rf_model, X_train_resampled, X_test, y_train_resampled, y_test, "rf_model_resampled.pkl")
# evaluate_and_save(svm_model, X_train_resampled, X_test, y_train_resampled, y_test, "svm_model_resampled.pkl")
# evaluate_and_save(nb_model, X_train_resampled, X_test, y_train_resampled, y_test, "nb_model_resampled.pkl")
# evaluate_and_save(knn_model, X_train_resampled, X_test, y_train_resampled, y_test, "knn_model_resampled.pkl")
```
